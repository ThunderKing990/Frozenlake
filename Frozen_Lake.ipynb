```python
import numpy as np
import gym
import random
import time
from IPython.display import clear_output
from colabgymrender.recorder import Recorder
```

```python
env = gym.make('FrozenLake-v1')
directory = './video'
env = Recorder(env, directory)
```

```python
action_size = env.action_space.n
state_size = env.observation_space.n
```

```python
qtable = np.zeros((state_size, action_size))
```

```python
num_episodes = 10000
max_steps_per_episode = 100

learning_rate = 0.1
discount_rate = 0.99

exploration_rate = 1
max_exploration_rate = 1
min_exploration_rate = 0.01
exploration_decay_rate= 0.004
```

```python
rewards_all_episodes = []
for episode in range(num_episodes):
    state = env.reset()
    done =False
    rewards_current_epsiode = 0
    
    for step in range(max_steps_per_episode):
        exploration_rate_threshold = random.uniform(0,1)
        
        if exploration_rate_threshold > exploration_rate:
            action = np.argmax(qtable[state,:])
        else:
            action = env.action_space.sample()
        
        new_state, reward, done, info = env.step(action)
        
        qtable[state, action] = qtable[state, action] * (1- learning_rate) + learning_rate * (reward + discount_rate * np.max(qtable[new_state, :]))

        state = new_state
        rewards_current_epsiode += reward

        if done == True:
          break
    #exploration_rate = min_exploration_rate + (max_exploration_rate - min_exploration_rate) * np.exp(-exploration_decay_rate * episode)
    exploration_rate *= 0.995
    rewards_all_episodes.append(rewards_current_epsiode)

rewards_per_thousand_epsiodes = np.split(np.array(rewards_all_episodes),num_episodes/1000)
count = 1000
print("********Average reward per thousand episodes********\n")
for r in rewards_per_thousand_epsiodes:
  print(f"{count} : {sum(r/1000)}")
  count += 1000

print("\n\n********Q Table********\n")
print(qtable)
```

```python
for episode in range(10):       
    state = env.reset()         
    done = False               
    print("*****EPISODE ", episode+1, "*****\n")            
    for step in range(max_steps_per_episode):   
      action = np.argmax(qtable[state,:])   
      new_state, reward, done, info = env.step(action)  
  
      if done == True :
        if reward == 1:
          print("****You reached the goal!****\n\n")
          env.play()   
          print("#----------------------------#\n\n")
        else:
          print("****You failed to reach the goal!****\n\n")
          env.play()     
          print("#----------------------------#\n\n")

        break
      state = new_state    
env.close()  
```
